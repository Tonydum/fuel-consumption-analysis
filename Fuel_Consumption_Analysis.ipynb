```python
# Assuming you've loaded the dataset into a DataFrame named fuel_data
# and the columns for engine size and combined fuel consumption are named 'ENGINE_SIZE' and 'COMB_L_100_km' respectively

# Calculate the Pearson correlation coefficient
correlation_coef, p_value = pearsonr(df['ENGINE SIZE'], df['COMB (L/100 km)'])

print(f'Pearson Correlation Coefficient: {correlation_coef}')
print(f'p-value: {p_value}')
```
